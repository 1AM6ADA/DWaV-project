```python
# Cell 2: Setup Ollama Connection (ИСПРАВЛЕННАЯ ВЕРСИЯ)

import subprocess
import sys
import time
import locale

def check_ollama_running():
    """Check if Ollama is running and available"""
    try:
        # Используем shell=True для Windows и правильную кодировку
        result = subprocess.run(
            'ollama list', 
            shell=True,
            capture_output=True, 
            text=True, 
            encoding='utf-8',  # Явно указываем UTF-8
            errors='ignore',    # Игнорируем ошибки кодировки
            timeout=10
        )
        return result.returncode == 0
    except Exception as e:
        print(f"Ollama check error: {e}")
        return False

def query_ollama(prompt, model="llama2", max_retries=2):
    """Send a query to Ollama and get response with proper encoding"""
    for attempt in range(max_retries):
        try:
            # Для Windows используем shell=True и правильную кодировку
            result = subprocess.run(
                f'ollama run {model} "{prompt}"',
                shell=True,
                capture_output=True,
                text=True,
                encoding='utf-8',  # Явно указываем UTF-8
                errors='ignore',    # Игнорируем ошибки кодировки
                timeout=45          # Увеличиваем таймаут
            )
            
            if result.returncode == 0:
                return result.stdout.strip()
            else:
                print(f"Attempt {attempt + 1}: Ollama returned error code {result.returncode}")
                if result.stderr:
                    print(f"Error: {result.stderr}")
                    
        except subprocess.TimeoutExpired:
            print(f"Attempt {attempt + 1}: Ollama timeout (45s)")
        except Exception as e:
            print(f"Attempt {attempt + 1}: Error: {e}")
        
    
    return None

print("Checking Ollama connection...")
if check_ollama_running():
    print("✅ Ollama is running")
    
    # Test Ollama with a simple query
    print("Testing Ollama with simple query...")
    test_response = query_ollama("Say 'Hello' in one word", max_retries=1)
    if test_response:
        print(f"✅ Ollama test successful: {test_response}")
    else:
        print("⚠️ Ollama test failed, but continuing...")
else:
    print("⚠️ Ollama is not running. Please start Ollama first:")
    print("   Run: ollama serve")
    print("   Then: ollama pull llama2")
```

```python
# Cell 3: Load the Dataset with Proper Error Handling

def load_wvs_data(file_path):
    """
    Load WVS trend file with proper handling of value labels
    The key fix: convert_categoricals=False to handle duplicate value labels
    """
    try:
        # First, check if file exists
        if not os.path.exists(file_path):
            print(f"❌ File not found: {file_path}")
            return None
            
        print(f"Attempting to load file: {file_path}")
        
        # IMPORTANT FIX: convert_categoricals=False prevents the value labels error
        df = pd.read_stata(file_path, convert_categoricals=False)
        
        print(f"✅ Dataset loaded successfully")
        print(f"   Shape: {df.shape}")
        print(f"   Rows: {len(df)}")
        print(f"   Columns: {len(df.columns)}")
        
        # Check data types
        object_cols = df.select_dtypes(include=['object']).columns.tolist()
        if object_cols:
            print(f"   Columns with object dtype: {len(object_cols)}")
            print(f"   First 5: {object_cols[:5]}")
        
        return df
        
    except Exception as e:
        print(f"❌ Error loading dataset: {e}")
        
        # Try alternative method with explicit encoding
        try:
            print("Trying alternative method with latin-1 encoding...")
            df = pd.read_stata(file_path, convert_categoricals=False, encoding='latin-1')
            print("✅ Successfully loaded with latin-1 encoding")
            return df
        except Exception as e2:
            print(f"❌ Alternative method also failed: {e2}")
            return None

# Specify your file path - using raw string to avoid Unicode escape error
file_path = r'C:\Users\HP\Documents\GitHub\DWaV-project\Trends_VS_1981_2022_Stata_v4_1.dta'

# Load the data
df = load_wvs_data(file_path)

# If loading fails, create sample data for demonstration
if df is None:
    print("\nCreating sample WVS data for demonstration...")
    np.random.seed(42)
    n_rows = 1000
    
    sample_data = {
        'S002': np.random.choice(range(1, 50), n_rows),
        'S003': np.random.choice(range(1960, 2023), n_rows),
        'S020': np.random.choice(range(1, 100), n_rows),
        'X001': np.random.choice([1, 2], n_rows),
        'X003': np.random.randint(18, 90, n_rows),
        'A001': np.random.choice([1, 2, 3, 4], n_rows),
        'A002': np.random.choice([1, 2, 3, 4], n_rows),
        'F114': np.random.choice([1, 2, 3, 4, 5], n_rows),
        'E069': np.random.choice([1, 2, 3, 4], n_rows),
        'E069_01': np.random.choice([1, 2, 3, 4], n_rows),
        'E179WVS': np.random.choice(['Party A', 'Party B', 'Party C'], n_rows),
    }
    
    df = pd.DataFrame(sample_data)
    print(f"✅ Sample data created with shape: {df.shape}")
```

```python
# Cell 5: Use Ollama to Decode Column Names (ИСПРАВЛЕННАЯ ВЕРСИЯ)

def decode_column_with_ollama(column_code):
    """
    Use Ollama to decode what a WVS column code means
    With better error handling and shorter prompts
    """
    
    # Более короткий промпт для уменьшения времени ответа
    prompt = f"""What does WVS column '{column_code}' mean? 
    Respond in this format:
    NAME: [variable name]
    DESCRIPTION: [what it measures]
    SCALE: [response format]
    Keep it brief."""
    
    try:
        response = query_ollama(prompt, max_retries=2)
        if response:
            return response
        else:
            # Fallback for common columns
            fallback_descriptions = {
                'studyno': "NAME: Study Number\nDESCRIPTION: Unique identifier for the study/wave\nSCALE: Numeric code",
                'version': "NAME: Dataset Version\nDESCRIPTION: Version number of the dataset\nSCALE: Version string",
                'doi': "NAME: Digital Object Identifier\nDESCRIPTION: DOI for citing the dataset\nSCALE: DOI string",
                'S002': "NAME: Country Code\nDESCRIPTION: Country identifier for the survey\nSCALE: Numeric country code",
                'S003': "NAME: ISO Country Code\nDESCRIPTION: Numeric ISO country code\nSCALE: 3-digit numeric code",
                'S020': "NAME: Wave Number\nDESCRIPTION: Survey wave identifier\nSCALE: Numeric wave number",
                'X001': "NAME: Gender\nDESCRIPTION: Respondent's gender\nSCALE: 1=Male, 2=Female",
                'X003': "NAME: Age\nDESCRIPTION: Respondent's age in years\nSCALE: Numeric age",
            }
            return fallback_descriptions.get(column_code, f"NAME: {column_code}\nDESCRIPTION: Unknown\nSCALE: Unknown")
            
    except Exception as e:
        print(f"Error decoding {column_code}: {e}")
        return f"NAME: {column_code}\nDESCRIPTION: Error in decoding\nSCALE: Unknown"

def decode_all_columns(df, sample_size=None):
    """
    Decode all column names using Ollama
    With better error handling and progress tracking
    """
    
    if not check_ollama_running():
        print("⚠️ Ollama not running. Using fallback naming.")
        return {col: f"NAME: {col}\nDESCRIPTION: Unknown\nSCALE: Unknown" for col in df.columns}
    
    columns_to_decode = df.columns.tolist()
    if sample_size and sample_size < len(columns_to_decode):
        columns_to_decode = columns_to_decode[:sample_size]
        print(f"🔄 Decoding first {sample_size} columns (sample mode)")
    else:
        print(f"🔄 Decoding all {len(columns_to_decode)} columns (this may take a while)")
    
    decoded_dict = {}
    total = len(columns_to_decode)
    
    for i, col in enumerate(columns_to_decode, 1):
        print(f"  Decoding {i}/{total}: {col}")
        
        try:
            decoded = decode_column_with_ollama(col)
            decoded_dict[col] = decoded
            print(f"    ✓ Got response for {col}")
        except Exception as e:
            print(f"    ✗ Failed for {col}: {e}")
            decoded_dict[col] = f"NAME: {col}\nDESCRIPTION: Decoding failed\nSCALE: Unknown"
        
        # Увеличиваем задержку между запросами
        time.sleep(2)
    
    return decoded_dict

# Decide how many columns to decode
decode_all = False  # Set to True to decode all columns

print("🔄 Starting column decoding with Ollama...")
if decode_all:
    column_decodings = decode_all_columns(df)
else:
    # Decode first 5 columns for testing
    column_decodings = decode_all_columns(df, sample_size=5)

print(f"\n✅ Decoded {len(column_decodings)} columns")
```

```python
# Cell 6: Parse Ollama Responses into Clean Descriptions

def parse_ollama_response(response, column_code):
    """
    Parse Ollama's response into a clean description
    """
    if not response:
        return column_code
    
    # Try to extract NAME, DESCRIPTION, SCALE
    name_match = re.search(r'NAME:\s*(.+)', response)
    desc_match = re.search(r'DESCRIPTION:\s*(.+)', response)
    scale_match = re.search(r'SCALE:\s*(.+)', response)
    
    parts = []
    if name_match:
        parts.append(name_match.group(1).strip())
    if desc_match:
        parts.append(desc_match.group(1).strip())
    
    if parts:
        return " | ".join(parts)
    else:
        # If parsing fails, return first line of response
        first_line = response.split('\n')[0].strip()
        return first_line if first_line else column_code

def create_column_mapping(decoded_dict):
    """
    Create a mapping from original column codes to clean descriptions
    """
    mapping = {}
    
    for code, response in decoded_dict.items():
        mapping[code] = parse_ollama_response(response, code)
    
    return mapping

# Create clean column mapping
column_mapping = create_column_mapping(column_decodings)

print("📋 Column mapping created:")
for code, desc in list(column_mapping.items())[:5]:
    print(f"  {code} -> {desc}")
```

```python
# Cell 7: Get Preprocessing Suggestions from Ollama

def get_preprocessing_suggestions(df):
    """
    Ask Ollama for preprocessing suggestions based on data structure
    """
    
    if not check_ollama_running():
        print("⚠️ Ollama not running, skipping suggestions")
        return None
    
    # Get sample of data
    col_sample = list(df.columns[:10])
    data_sample = df.head(3).to_string()
    
    prompt = f"""You are analyzing the World Values Survey dataset.
    
Column names (first 10): {col_sample}

First 3 rows of data:
{data_sample}

Based on this WVS data structure, suggest 5 specific preprocessing steps I should perform.
Focus on:
- Handling missing values (WVS uses -5, -4, -3, -2, -1 for missing)
- Data type optimization
- Creating derived variables
- Handling country-specific codes
- Preparing for analysis

Provide each step as a clear, actionable item."""
    
    print("🤖 Getting preprocessing suggestions from Ollama...")
    suggestions = query_ollama(prompt, model="llama2", max_retries=2)
    
    if suggestions:
        print("\n📋 Ollama's preprocessing suggestions:")
        print("-" * 50)
        print(suggestions)
        return suggestions
    else:
        print("Could not get suggestions from Ollama")
        return None

preprocessing_suggestions = get_preprocessing_suggestions(df)
```
